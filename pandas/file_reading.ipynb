# 二、hdf5文件

```python
my_hdf5 = pd.read_hdf('./day3资料/02-代码/stock_data/day/day_close.h5') # 单个键时key= 可以省略
my_hdf5.head()
```

```python
my_hdf5.to_hdf('my_hdf5.h5', key='close') # 存储hdf5文件，key一定要有
```

```python
pd.read_hdf('my_hdf5.h5', key='close').head() # 单个键时key= 可以省略
```

```python
my_hdf5_2 = pd.read_hdf('./day3资料/02-代码/stock_data/day/day_open.h5')
my_hdf5_2.head()
```

```python
my_hdf5_2.to_hdf('my_hdf5.h5', key='open') # 将day_open的数据存储为另一个key
```

```python
pd.read_hdf('my_hdf5.h5', key='open').head() # 多个键时必须指定key= 
```
